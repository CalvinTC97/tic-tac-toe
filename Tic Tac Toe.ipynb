```python
from IPython.display import clear_output
from random import randint

exampleboard = ['ignore','1','2','3','4','5','6','7','8','9']
myboard = ['ignore',' ',' ',' ',' ',' ',' ',' ',' ',' ']

playerx_turn = True
first_round = True
game_playing = True
game_over = False

def display_board(myboard):
    print(myboard[7] + '|' + myboard[8] + '|' + myboard[9])
    print('-----')
    print(myboard[4] + '|' + myboard[5] + '|' + myboard[6])
    print('-----')
    print(myboard[1] + '|' + myboard[2] + '|' + myboard[3])

def change_board(myboard):
    choice_verified = False
    
    while choice_verified == False:
        myinput = input("Which tile do you want to play? ")
        if (myinput.lower() == 'quit'):
            return False
        try: myinput = int(myinput)
        except: ValueError
        if (isinstance(myinput,int) == False):
            print("Please enter a number between 1 and 9 ")
        elif (int(myinput) > 9):
            print("Please enter a number between 1 and 9 ")
        elif myboard[myinput] != " ":
            print("That tile is already taken")
        else:
            myinput = int(myinput)
            choice_verified = True
    
    if (playerx_turn == True):
        myboard[myinput] = 'x'
    
    if (playerx_turn == False):
        myboard[myinput] = 'o'
    
    clear_output()
    return display_board(myboard)
    choice_verified = False

def check_board(myboard, game_over):
    for tile in [4,5,6]:
        if ((myboard[tile] == myboard[tile+3] == myboard[tile-3]) and (myboard[tile] != " ")):
            game_over = True
            return game_over
    
    for tile in [2,5,8]:
        if ((myboard[tile] == myboard[tile+1] == myboard[tile-1]) and (myboard[tile] != " ")):
            game_over = True
            return game_over
    
    if ((myboard[7] == myboard[5] == myboard[3]) and (myboard[5] != " ")):
        game_over = True
        return game_over
    
    if ((myboard[1] == myboard[5] == myboard[9]) and (myboard[5] != " ")):
        game_over = True
        return game_over
    
    if ((" " in myboard) != True):
        game_over = True
        return "NA"

    
print("Welcome to Tic Tac Toe, the most strategic game you can play with the least complexity/motor skills! \n\nPlayers will take turns entering an 'x' or an 'o', with the first player being randomly chosen. \nTo select a tile, press the corresponding number on your keypad:")
display_board(exampleboard)
print("Enter 'Quit' to exit at any time. Good luck! \n\n\n")


while game_playing == True:
    if (first_round == True):
        myboard = ['ignore',' ',' ',' ',' ',' ',' ',' ',' ',' ']
        if (randint(1,2) == 1):
            print('Player x will go first')
            playerx_turn = True
            first_round = False
        else:
            print('Player o will go first')
            playerx_turn = False
            first_round = False        
    
    if (change_board(myboard) == False):
        print('Goodbye')
        game_playing = False
        break
    
    game_over = check_board(myboard, game_over)
    
    if (game_over == "NA"):
        print("Draw! Either you're both really good or really bad")
        
        answer = input("Do you want to play again? Yes / No ")
        if (answer.lower() == 'yes'):
            game_over = False
            first_round = True
        else:
            print('Goodbye')
            game_playing = False
        
    elif (game_over == True):
        if (playerx_turn == True):
            print('Player x wins! X-cellent work!')
        elif (playerx_turn == False):
            print('Player o wins! O-mazing job!')
        
        answer = input("Do you want to play again? Yes / No ")
        if (answer.lower() == 'yes'):
            clear_output()
            game_over = False
            first_round = True
        else:
            print('Goodbye')
            game_playing = False

            
    if ((playerx_turn == True) and (game_playing == True)):
        playerx_turn = False
        print("\n***************")
        print("PLAYER O'S TURN")
        print("***************")
    elif ((playerx_turn == False) and (game_playing == True)):
        playerx_turn = True
        print("\n***************")
        print("PLAYER X'S TURN")
        print("***************")
```
